### 비지도 변환
<pre>
비지도 변환은 데이터를 새롭게 표현하여 사람이나 다른 머신러닝 알고리즘이 원래 데이터보다 쉽게 해석할 수 있도록 만드는 알고리즘이다.
</pre>
<pre>
비지도 변환이 널리 사용되는 분야는 특성이 많은 고차원 데이터의 특성(feature)의 수를 줄이면서 꼭 필요한 특징을 포함한 데이터로 표현하는 방법인 차원 축소(dimensionality reduction)이다.
</pre>

### 예 : 텍스트 문서에서 주제를 추출하는 것.

## 군집 알고리즘(Clustering)
<pre>
데이터를 비슷한 것끼리 그룹으로 묶는 것.
</pre>

## 비지도학습에서 어려운 점.
<pre>
(가) 알고리즘이 뭔가 유용한 것을 학습했는지 평가하는 것.
(나) 알고리즘에게 우리가 뭔가 원하는 것을 알려줄 수 없다.
==> 이런 이유로 비지도 학습 알고리즘은 데이터 과학자가 데이터를 잘 이해하고 싶을 때 탐색적 분석 단계에서 많이 사용.
</pre>


### 3-3 데이터 전처리와 스케일 조정

```python
import mglearn
%matplotlib inline
```

```python
mglearn.plots.plot_scaling()
```

#### (가) StandardScaler
<pre>
(가) 각 특성(feature)의 평균을 0, 분산을 1로 변경
(나) 이 방법은 feature(특성)의 최솟값과 최댓값 크기를 제한하지 않는다.
</pre>

#### (나) RobustScaler
<pre>
(가) 같은 스케일을 갖는다. StandardScaler과 비슷
(나) 이 방법은 평균과 분산 대신 중간 값(median)과 사분위 값(quantile)을 사용.
--> 전체 데이터와 아주 동떨어진 데이터 포인트에 영향을 받지 않음.
</pre>

#### (다) MinMaxScaler
<pre>
(가) 모든 특성이 정확하게 0과 1사이에 위치하도록 데이터를 변경.
</pre>

#### (라) Normalizer
<pre>
(가) 유클리디안 길이가 1이 되도록 데이터 포인트를 조정.
(나) 지름이 1인 원에 데이터 포인트를 투영한다.
--> 정규화는 특성 벡터의 길이는 상관없고 데이터의 방향(또는 각도)만이 중요할 때 많이 사용.
</pre>

### 3-4-1  주성분 분석(PCA)
<pre>
주성분 분석은 특성들이 통계적으로 상관관계가 없도록 데이터셋을 회전시키는 기술이다.
</pre>

```python
mglearn.plots.plot_pca_illustration()
```

### 첫번째 그래프
<pre>
(가) PCA 알고리즘은 '성분1'이라고 쓰여 있는, 분산이 가장 큰 방향을 찾는다.
--> 데이터에서 가장 많은 정보를 담고 있는 방향
--> 이 방향이 특성들의 상관관계가 가장 큰 방향

(나) 이 알고리즘은 첫 번째 방향과 직각인 방향 중에서 가장 많은 정보를 담는 방향을 찾는다.
--> 2차원에서는 하나
--> 고차원에서는 (무한히) 많은 직각 방향이 있을 수 있다.

(다) 화살표의 머리와 꼬리는 의미가 없다.

(라) 이런 과정을 거쳐 찾은 방향의 데이터에 있는 주된 분산의 방향이라고 해서 주성분이라 한다.
(라) 일반적으로 원본 특성 개수만큼의 주성분이 있다.
</pre>

### 두번째 그래프
<pre>
(가) 주성분1과 2를 각각 x축과 y축에 나란하도록 회전.
(나) 회전하기 전에 평균을 빼서 중심을 원점에 맞춤
</pre>

### 세번째 그래프
<pre>
(가) 세번째 그래프는 첫 번째 주성분만 있다. 2차원 -> 1차원
--> 원본 특성중 하나만 남기는 것이 아닌 가장 유용한 방향을 찾은 주성분1만 유지.
</pre>

### 네번째 그래프
<pre>
(가) 데이터에 평균을 더해서 반대로 회전.
(나) 이 데이터 포인트들은 원래 특성 공간에 있지만 첫번재 주성분만 담고 있다.
</pre>
