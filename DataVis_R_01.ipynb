## 01. Corr 확인하기

```r
# install.packages("corrplot")
library(corrplot)
```

```r
numericVars <- which(sapply(df, is.numeric))  ## 정수형 확인 변수 선택
numericVarNames <- names(numericVars)         ## 변수의 이름 확인
numericVarNames
cat('There are', length(numericVars), 'numeric variables')
```

```r
head(df[, numericVars] )  # 선택된 열의 값을 확인
dim(df[, numericVars])
```

```r
all_numVar <- df[, numericVars]
cor_numVar <- cor(all_numVar, use="pairwise.complete.obs") # correlations of all numeric variables
# pairwise.complete.obs : 상관계수가 계산되는 변수들만을 대상으로 결측값이 있는 case 제거한  상관계수 계산
```

#### SalePrice와 함께 상관관계를 내림차순 정렬함. [ sort on decreasing correlations with SalePrice ]

```r
cor_sorted <- as.matrix(sort(cor_numVar[,'SalePrice'], decreasing = TRUE))
cor_sorted
```

#### 상관계수가 0.5이상만 선택합니다.

```r
mulcor = apply(cor_sorted, 1, function(x) abs(x)>0.5)   # 1:행 적용, 2:열 적용
mulcor
```

```r
which(mulcor)
```

```r
# select only high corelations
CorHigh <- names(which(mulcor))
CorHigh
cor_numVar <- cor_numVar[CorHigh, CorHigh]
cor_numVar
```

```r
corrplot.mixed(cor_numVar, tl.col="black", tl.pos = "lt")
```

### 소스 코드 정리

```r
library(corrplot)
numericVars <- which(sapply(df, is.numeric))  ## 정수형 확인 변수 선택
numericVarNames <- names(numericVars)         ## 변수의 이름 확인
cat('There are', length(numericVars), 'numeric variables')
dim(df[, numericVars])

all_numVar <- df[, numericVars]   # 정수형 변수 열을 선택
cor_numVar <- cor(all_numVar, use="pairwise.complete.obs")  # 상관계수(결측치 열 제외)

# 'SalePrice'열의 상관계수의 값을 내림차순 정렬
cor_sorted <- as.matrix(sort(cor_numVar[,'SalePrice'], decreasing = TRUE)) 

# 상관계수가 0.5 이상의 행을 true, 아니면 false 처리
mulcor = apply(cor_sorted, 1, function(x) abs(x)>0.5)   # 1:행 적용, 2:열 적용

# 높은 친구 선택
CorHigh <- names(which(mulcor))
cor_numVar <- cor_numVar[CorHigh, CorHigh]  # 행은 0.5이상 행 선택, 열도 0.5이상인 열 선택
cor_numVar

# 그래프 그리기
corrplot.mixed(cor_numVar, tl.col="black", tl.pos = "lt")
```

## 02. 범주형 데이터 시각화하기

```r
train <- read.csv("D:/dataset/House_Prices/train.csv")
test <- read.csv("D:/dataset/House_Prices/test.csv")
```

```r
require(ggplot2) # for data visualization
```

```r
# categoric data(범주형 데이터)를 시각화 하기 위한 Function(함수)
plot.categoric <- function(cols, df){
  for (col in cols) {
    order.cols <- names(sort(table(df.combined[,col]), decreasing = TRUE))
  
    num.plot <- qplot(df[,col]) +
      geom_bar(fill = 'cornflowerblue') +
      geom_text(aes(label = ..count..), stat='count', vjust=-0.5) +
      theme_minimal() +
      scale_y_continuous(limits = c(0,max(table(df[,col]))*1.1)) +
      scale_x_discrete(limits = order.cols) +
      xlab(col) +
      theme(axis.text.x = element_text(angle = 30, size=12))
  
    print(num.plot)
  }
}
```

#### 데이터 셋 합치기

```r
# combine the datasets
df.combined <- rbind(within(train, rm('Id','SalePrice')), within(test, rm('Id')))
```

## 03. 결측치 확인 시각화하기

```r
# install.packages("Amelia")
library(Amelia)
```

```r
train01 <- read.csv("D:/dataset/House_Prices/train.csv")
test01 <- read.csv("D:/dataset/House_Prices/test.csv")
```

```r
names(train01)
names(test01)
```

```r
a = within(train01, rm('Id','SalePrice'))
b = within(test01, rm('Id'))
data_raw <-  rbind(a, b)

train01$logSalePrice <- log(train01$SalePrice) ## create log of sales price
```

```r

## MSSubClass 
data_raw$MSSubClass <- as.factor(data_raw$MSSubClass)
data_raw$MoSold <- as.factor(data_raw$MoSold)
data_raw$YrSold <- as.factor(data_raw$YrSold)
```

```r
## NA as a category: without 
var_NAtoWithout<-c("Alley","BsmtQual","BsmtCond","BsmtExposure","BsmtFinType1","BsmtFinType2","FireplaceQu","GarageType","GarageFinish","GarageQual","GarageCond","PoolQC","Fence","MiscFeature")
## impute NA to without
without<-function(data,var){
        levels(data[,var]) <- c(levels(data[,var]), "without")
        data[,var][is.na(data[,var])] <- "without"
        return(data[,var])
}
data_combined<-data_raw
for (i in 1:length(var_NAtoWithout)){
       data_combined[,var_NAtoWithout[i]]<-without(data_raw,var_NAtoWithout[i]) 
}
```

```r
missmap(data_raw[-1], col=c('Red', 'Green'), y.cex=0.2, x.cex=0.8,main = "Before Imputation")
```

```r
## Plot NA pattern 
missmap(data_combined[-1], col=c('Red', 'Green'), y.cex=0.2, x.cex=0.8,main = "After Imputation")
```
